# Data Processing and Cleaning

## Type casting - Date Column and duration column

````python
dates=df['date_added']
````

````python
dates=pd.to_datetime(dates,format='%B %d, %Y',errors='coerce')
df['dates']=dates
df['dates']
````

````python
# Removing the strings from duration column

# Method 1

df['duration']=df['duration'].apply(lambda x: str(x).split(' ')[0])


'''
Method 2
dur=df['duration']
dur.str.replace(' min','')
dur.str.replace(' mins','')
dur.str.replace(' Seasons','')
dur.str.replace(' Season','')

'''
````

## Unnesting of Cast,director,listen_in(genres),country

````python
# so now we have to unnest the values in cast,listed_in column
tem=df['cast'].apply(lambda x: str(x).split(', ')).tolist()

````

````python
df_cast=pd.DataFrame(tem,index=df['title'])

df_cast
````

````python
#Let's the stack the df_xast
df_cast=df_cast.stack()

#Lets reset the index as well
df_cast=df_cast.reset_index()

# let's drop the level_1 column
df_cast.drop('level_1',axis=1,inplace=True)

# After that rename 0 column to cast
df_cast.rename(columns={0:'cast'},inplace=True)
df_cast


````

````python
# Let's filter out the data for ranbir kapoor movies

df_cast[df_cast['cast']=='Ranbir Kapoor']
````

````python
# Lets check the top 10 cast
df_cast['cast'].value_counts().head(10)
````

````python
# let's create the function for the step i perform for title and cast column

def processing(df,column):
  tem=df[column].apply(lambda x: str(x).split(', ')).tolist()
  df_temp=pd.DataFrame(tem,index=df['title']).stack().reset_index()
  df_temp.drop('level_1',axis=1,inplace=True)
  df_temp=df_temp.rename(columns={0:column})
  return df_temp
````

````python
# We have to do the same for director column
df_director=processing(df,'director')
df_director
````

````python
# Top 10 Directors Country Wise
df_director['director'].value_counts().head(10)
````

````python
df_country=processing(df,'country')
df_country
````

````python
# Top 10 Countries
df_country['country'].value_counts().head(10)
````

````python
df_genre=processing(df,'listed_in')
df_genre
````

````python
df_genre['listed_in'].unique()
````

````python
# Let's create a new dataset by merging df_cast,df_director,df_country,df_genre

df_new=df_cast.merge(df_director,on='title').merge(df_genre,on='title').merge(df_country,on='title')
df_new
````

````python
'''
Filling null values with the most frequent value (like "United States" in this case) is a common technique called
mode imputation. However, it's not always the best practice as it can introduce bias and
distort the data distribution.
'''
df_country['country'].mode()
#df2['country'].fillna(df2['country'].mode()[0], inplace=True)
````

# Final Dataset

````python
'''

Now we have five data set here :
Original Datset -- df
For Cast -- df_cast
For Directors -- df_director
For Genres -- df_listed_in
For Countries -- df_country

We have to merge all these data set based on the title column and create one final dataset and then have to manage the null values

'''
df_final=df[['show_id','title','type','dates','release_year','rating','duration']]
df_final=df_final.merge(df_cast,on='title',how='left').merge(df_director,on='title',how='left').merge(df_genre,on='title',how='left').merge(df_country,on='title',how='left')
df_final
````

````python
df_final['country'] = df_final['country'].replace('nan', np.nan)
df_final['country'] = df_final['country'].replace('', np.nan)
df_final['cast']=df_final['cast'].replace('nan',np.nan)
df_final['director']=df_final['director'].replace('nan',np.nan)
df_final['listed_in']=df_final['listed_in'].replace('nan',np.nan)


````

````python
df_final.isnull().sum()
````

## Handling Null Values

````python
top_cast=dict(df_final.groupby('country')['cast'].apply(lambda x: x.value_counts().index[0] if len(x.value_counts()) > 0 else None))
top_cast
````

````python
# Now We have to fill the null values let's start for Cast column

top_directors=dict(df_final.groupby('country')['director'].apply(lambda x: x.value_counts().index[0] if len(x.value_counts()) > 0 else None))
top_directors

````

````python
top_country=df_final['country'].mode()
top_country
````

````python
# Let fill null values for country column with country with the top country

df_final['country'].fillna(top_country[0],inplace=True)
````

````python
df_final['cast'].fillna(df_final['country'].map(top_cast),inplace=True)
````

````python
df_final['director'].fillna(df_final['country'].map(top_directors),inplace=True)
````

````python
df_final.isnull().sum()*100/df_final.shape[0]
````

````python
# Let's Drop the null values for the column where it is less than 1 percent as it not have any major impact due to very low number
df_final.dropna(subset=['cast','director','dates','rating'],inplace=True)
````

````python
#Lets again the check the null values
df_final.isnull().sum()*100/df_final.shape[0]
````

````python
# As we don't have any null values in the data frame lets go ahead and do the analysis
df_final.rename(columns={'listed_in':'genre'},inplace=True)
df_final.head(5)
````

## Best month to release movie/tv show

````python
# Find the Best month to release a movie/tv show
df_date=df
df_date['date_added'] = pd.to_datetime(df_date['date_added'], errors='coerce')
df_date['month'] = df_date['date_added'].dt.month_name()
top_month=df_date.groupby(['month','type'])['title'].count().unstack()
order=['January','February','March','April','May','June','July','August','September','October','November','December']
top_month=top_month.reindex(order,axis=0)
top_month.fillna(0,inplace=True)
top_month
````
